## Passing input to Actor

```python
import json # Wichtig: Modul f√ºr JSON-Handling importieren
from apify_client import ApifyClient

# Dein Token
TOKEN = ''

def main() -> None:
    client = ApifyClient(token=TOKEN)
    ACTOR_ID = 'XtaWFhbtfxyzqrFmd'

    # --- KONFIGURATION ---
    keyword = "fitness"
    country = "DE"
    limit = 10 # F√ºr den Mock reichen 10-20 Items v√∂llig aus
    # ---------------------

    print(f"üõ†Ô∏è Baue Such-URL f√ºr '{keyword}' in '{country}'...")

    search_url = (
        f"https://www.facebook.com/ads/library/"
        f"?active_status=all"
        f"&ad_type=all"
        f"&country={country}"
        f"&q={keyword}"
        f"&sort_data[direction]=desc&sort_data[mode]=relevancy_monthly_grouped"
        f"&media_type=all"
    )

    run_input = {
        "urls": [{"url": search_url}], 
        "maxItems": limit,
    }

    print(f"üöÄ Starte Scraper ({ACTOR_ID})...")

    try:
        run = client.actor(ACTOR_ID).call(run_input=run_input, timeout_secs=120)

        if run:
            dataset_id = run['defaultDatasetId']
            print(f"‚úÖ Fertig! Dataset ID: {dataset_id}")
            
            # Daten laden
            dataset_items = client.dataset(dataset_id).list_items().items
            print(f"üìä Gefundene Anzeigen: {len(dataset_items)}")
            
            # --- NEU: Speichern in Datei ---
            filename = "mock_data.json"
            print(f"üíæ Speichere Daten in '{filename}'...")
            
            with open(filename, "w", encoding="utf-8") as f:
                # indent=4 macht es sch√∂n lesbar, ensure_ascii=False erh√§lt Emojis/Sonderzeichen
                json.dump(dataset_items, f, indent=4, ensure_ascii=False)
                
            print(f"‚úÖ Erfolgreich gespeichert! Du kannst die Datei '{filename}' jetzt √∂ffnen.")
            # -------------------------------

        else:
            print("‚ùå Keine Daten zur√ºckbekommen.")

    except Exception as e:
        print(f"‚ùå Ein Fehler ist aufgetreten: {e}")

if __name__ == '__main__':
    main()
```
